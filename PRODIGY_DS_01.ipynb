```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data_set = pd.read_csv("population.csv")
data_set.head(250)
```

```python
countries_to_plot=["United States","Canada","Brazil","Japan","Germany","France"]
```

```python
filtered_data=data_set[data_set['Country Name'].isin(countries_to_plot)]
```

```python
plt.bar(filtered_data['Country Name'],filtered_data['2021'])
plt.xlabel('Countries')
plt.ylabel('Population in 2021')
plt.title('Population of selected countries in 2021')
plt.show()
```
